### Parte 2: Ensamblaje de novo mediante Velvet

Velvet es un ensamblador de genomas dise침ado espec칤ficamente para secuencias cortas generadas por tecnolog칤as de secuenciaci칩n de alto rendimiento, como Illumina. Fue desarrollado por Daniel Zerbino y Ewan Birney en el EMBL-EBI y publicado en 2008.

游닄Publicaci칩n: https://europepmc.org/article/pmc/2336801

**Caracter칤sticas principales:**
- Basado en grafos de de Bruijn: Velvet transforma las lecturas en un grafo de de Bruijn, lo que permite ensamblar eficientemente secuencias cortas, incluso en presencia de errores.

- Optimizado para lecturas cortas: Ideal para lecturas de entre 35 y 300 pb, que eran comunes en las primeras generaciones de secuenciaci칩n masiva.

**Componentes clave:**
- velveth: construye el grafo de de Bruijn a partir de las lecturas.

- velvetg: refina el grafo, elimina errores, resuelve ambig칲edades y genera los contigs ensamblados.


**Ventajas:**
- R치pido y eficiente en recursos para su 칠poca.

- Permite integrar pares de extremos (paired-end reads) para mejorar el ensamblaje.

- Ofrece varias opciones para limpieza de errores y ajuste de par치metros de ensamblaje.


**Limitaciones:**
- Su rendimiento disminuye con lecturas m치s largas o datos m치s ruidosos.

- No es ideal para ensamblajes altamente complejos como metagenomas o genomas muy repetitivos (aunque hay extensiones como MetaVelvet para metagen칩mica).

- 쯈u칠 funci칩n tienen los argumentos -exp_cov auto y -cov_cutoff auto ?

**R/** Al usar auto en estas dos opciones permite que el velvetg decida automaticaente los valores que considere mas optimos segun la distrbucion de k-mers qu velveth genero en el paso 1.

**-exp_cov auto** (Expected K-mer Coverage), le indica a velvetg cual es la cobertura promedio que se espera para los k-mers que pertenecen al genoma real.

**-cov_cutoff auto** (Coverage Cutoff), establece un umbral minimo de cobertura para que un k-mer sea considerado en el ensamblaje. Elimina k-mers con baja frecuencia.

### Parte 3: Ensamblaje de novo mediante MegaHit

**MEGAHIT** es un ensamblador de genomas dise침ado para manejar grandes vol칰menes de datos de secuenciaci칩n, especialmente (aunqe no limitado) aquellos provenientes de estudios metagen칩micos, donde se busca ensamblar m칰ltiples genomas microbianos a partir de muestras ambientales complejas.


**Caracter칤sticas principales:**

- Basado en grafos de de Bruijn comprimidos: Utiliza una estructura de datos eficiente llamada succinct de Bruijn graph (SdBG), que permite representar el grafo con un uso m칤nimo de memoria.

- Dise침ado para escalar bien: Capaz de ensamblar datasets muy grandes, incluso en computadoras con recursos moderados, gracias a su enfoque optimizado de uso de memoria y CPU.

- Soporte para secuencias cortas y largas: Aunque fue optimizado inicialmente para lecturas cortas (Illumina), versiones m치s recientes permiten trabajar tambi칠n con lecturas largas (por ejemplo, Nanopore o PacBio) mediante estrategias h칤bridas.


**Componentes clave:**

- Monol칤tico: todo el proceso de ensamblaje est치 integrado en un solo comando (megahit), simplificando su uso.

- Permite ensamblajes de novo, sin necesidad de genoma de referencia.

- Integra mecanismos de limpieza de errores y construcci칩n jer치rquica del grafo (multi-k-mer), lo que mejora la calidad del ensamblaje.


**Ventajas:**

- Altamente eficiente en memoria y velocidad, incluso con muestras metagen칩micas complejas.

- Requiere pocos pasos y es f치cil de usar.

- Bien mantenido y ampliamente adoptado en estudios metagen칩micos.


**Limitaciones:**

- Aunque muy robusto, en algunos casos puede generar ensamblajes fragmentados, especialmente en presencia de organismos con genomas muy similares.

- Menos adecuado para ensamblajes donde se requiera informaci칩n estructural detallada (como reordenamientos o ensamblajes haplot칤picos), ya que su enfoque se centra en la eficiencia y escalabilidad.

- 쮺칩mo podria optimizar el tiempo de an치lisis de megahit para genomas o metagenomas de gran tama침o?

- meta-sensitive: Similar a la configuraci칩n por defecto. Bueno para metagenomas de complejidad media, pero puede ser lento para conjuntos de datos muy grandes.
- meta-large: Recomendado para metagenomas grandes y complejos, ya que usa kmers mas espaciados, reduciendo complejidad y tiempo de ejecucion
- meta huge: Para ensamblajes de metagenomas extremadamente grandes (Gbp o Tbp)

- Ajuste de hilos: megahit -t 16 (16 hilos)
- Ajuste de memoria: megahit -m 0.8 (80% de RAM)
